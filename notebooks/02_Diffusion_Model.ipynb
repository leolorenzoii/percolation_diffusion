# Diffusion Model

```python
from matplotlib import rcParams
import matplotlib.pyplot as plt

rcParams.update({'figure.figsize': (8, 6), 'axes.spines.top': False,
                 'axes.spines.right': False, 'axes.labelsize': 14,
                 'axes.titlesize': 16, 'axes.titleweight': 'bold',
                 'lines.linewidth': 3})
```

## No Recovery

```python
# Set model parameters
N = 1000
c = 0.10
tau = 0.05

# Set initial number of infected people
x_0 = 5
```

```python
# Set time range
time_range = 2000
x_t = [x_0]

# Generate the number of people infected at given time t
for t in range(time_range):
    x_t.append(x_t[-1] + N*c*tau*(x_t[-1] / N)* ((N - x_t[-1]) / N))
```

```python
# Plot results
fig, ax = plt.subplots()
ax.plot(x_t)
ax.set_xlabel("Time")
ax.set_ylabel("Number of People Infected");
```
